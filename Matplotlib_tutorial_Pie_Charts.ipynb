```python
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
```

```python
slices = [40,60]
plt.pie(slices)
plt.title('My Pie Chart')
```

```python
labels = ['Sixty','Forty']
plt.pie(slices, labels = labels)
plt.title('My Pie Chart')
```

```python
# We can also add colors with the name of the colors that is built-in, or by using the hex_values.
slices = [40,60,30,50]
labels = ['Sixty','Forty','Extra1','Extra2']
color = ['m','b','r','y']
plt.pie(slices, labels = labels, colors = color)
plt.title('My Pie Chart')
```

```python
# We can add edgecolors to avoid jumping up among slices.
plt.pie(slices, labels = labels, colors = color, wedgeprops = {'edgecolor':'black'})
plt.title('My Pie Chart')
```

```python
# Language Popularity
slices_2 = [59219, 55466, 47544, 36443, 35917,
          31991, 27097, 23030, 20524, 18523,
          18017, 7920, 7331, 7201, 5833]

labels_2 = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java',
          'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript',
          'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices_2, labels = labels_2)
plt.title('My Pie Chart')
plt.tight_layout()
```

## Note:


> It is not good to use pie charts with a large data, it will get extremely crowded and it might take us some time to find what we are looking for.
> We don"t really prefer to use pie charts when we have more that 5 items. we can use bar charts instead.

```python
labels_2 = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
slices_2 = [59219, 55466, 47544, 36443, 35917]
plt.style.use('fivethirtyeight')
plt.pie(slices_2, labels = labels_2)
plt.title('My Pie Chart')
plt.tight_layout()
```

### We can explode some items we needfrom the chart by using { explode } argument
> It will contain float values, and it represents the fraction of te radius that we'd like to explode. 

```python
labels_2 = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
slices_2 = [59219, 55466, 47544, 36443, 35917]
explode = [0,0,0,0.1,0]

# We can also add shadow of the slices in the pie charts. 
# Adding the percentage values.
plt.pie(slices_2, labels = labels_2, explode = explode, shadow = True, autopct = '%1.1f%%')
plt.title('My Pie Chart')
plt.tight_layout()
```
